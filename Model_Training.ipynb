```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

file_path = "/teamspace/studios/this_studio/1/02-14-2018.csv"
df = pd.read_csv(file_path)

cols = [
    'Dst Port', 'Protocol', 'Flow Duration',
    'Tot Fwd Pkts', 'Tot Bwd Pkts',
    'TotLen Fwd Pkts', 'TotLen Bwd Pkts',
    'Fwd Pkt Len Mean', 'Bwd Pkt Len Mean',
    'Flow Byts/s', 'Flow Pkts/s',
    'Pkt Len Mean', 'Pkt Len Std',
    'Fwd Header Len', 'Bwd Header Len',
    'Fwd Pkts/s', 'Bwd Pkts/s',
    'SYN Flag Cnt', 'ACK Flag Cnt', 'RST Flag Cnt',
    'Active Mean', 'Idle Mean', 'Label'
]

df = df[cols].dropna()

label_enc = LabelEncoder()
df['Label'] = label_enc.fit_transform(df['Label'])

X = df.drop(columns=['Label']).values
y = df['Label'].values

X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)

unique, counts = np.unique(y.numpy(), return_counts=True)
print("Label distribution:")
for lbl, cnt in zip(unique, counts):
    print(f"  {label_enc.inverse_transform([lbl])[0]}: {cnt}")

class IDSDataset(Dataset):
    def __init__(self, X, y):
        self.X, self.y = X, y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = IDSDataset(X, y)
n_total = len(dataset)
train_size = int(0.7 * n_total)
val_size = int(0.2 * n_total)
test_size = n_total - train_size - val_size

train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_dl = DataLoader(train_ds, batch_size=64, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=64)
test_dl = DataLoader(test_ds, batch_size=64)

class TransformerIDS(nn.Module):
    def __init__(self, input_dim, n_classes, d_model=64, nhead=4, num_layers=2):
        super().__init__()
        self.embed = nn.Linear(input_dim, d_model)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model, nhead, dim_feedforward=128, dropout=0.1, activation='relu'
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers)
        self.fc = nn.Linear(d_model, n_classes)

    def forward(self, x):
        x = self.embed(x).unsqueeze(0)
        x = self.transformer(x)
        x = x.mean(0)  
        return self.fc(x)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = TransformerIDS(X.shape[1], len(np.unique(y))).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

print("\nTraining started...\n")
for epoch in range(10):
    model.train()
    total_loss = 0
    for xb, yb in train_dl:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1:02d}: loss={total_loss/len(train_dl):.4f}")

def evaluate(dl):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for xb, yb in dl:
            xb, yb = xb.to(device), yb.to(device)
            preds = model(xb).argmax(1)
            correct += (preds == yb).sum().item()
            total += len(yb)
    return correct / total if total > 0 else 0

print("\nVal Acc:", evaluate(val_dl))
print("Test Acc:", evaluate(test_dl))
```
